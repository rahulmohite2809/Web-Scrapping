# Web Scrapping : IMDb Website (Movies Rating)

```python
# Import Libraries:
from bs4 import BeautifulSoup
import requests
import openpyxl
```

```python
#Getting Data:
Source=requests.get("https://www.imdb.com/chart/top/")
```

```python
#Beautiful Soup:
Soup=BeautifulSoup(Source.text,"html.parser")
```

```python
#finding all data having class="lister-list" 
Movies=Soup.find("tbody", class_="lister-list")
#print(Movies)
```

```python
#Inside class_="lister-list" find all "tr" tags
Movies=Soup.find("tbody",class_="lister-list").find_all("tr")
print(len(Movies))
```

```python
# Using For loop finding data of all Rank:
for movie in Movies:
    Rank=movie.find("td",class_="titleColumn").get_text(strip=True).split(".")[0]
    print(Rank)
    
    
```

```python
# Using For loop finding data of all Rank Name :
for movie in Movies:
    Name=movie.find("td",class_="titleColumn").a.text
    print(Name)
    
    
```

```python
# Using For loop finding data of all Year:
for movie in Movies:
    Year=movie.find("td",class_="titleColumn").span.text.strip("()")
    print(Year)
    
    
```

```python
## Using For loop finding data of all Rating:

for movie in Movies:
    Rating=movie.find("td",class_="ratingColumn imdbRating").strong.text
    print(Rating)
    
```

Create Excel File to Store the Data

```python
#create excel Workbook:
excel=openpyxl.Workbook()

sheet=excel.active
sheet.title="Top Rated Movies"
sheet.append(["Rank","Name","Year","Rating"])
```

```python
#Add all info in one loop and append into sheet :

for movie in Movies:
    Rank=movie.find("td",class_="titleColumn").get_text(strip=True).split(".")[0]
    Name=movie.find("td",class_="titleColumn").a.text
    Year=movie.find("td",class_="titleColumn").span.text.strip("()")
    Rating=movie.find("td",class_="ratingColumn imdbRating").strong.text
   # print(Rank,Name,Year,Rating)
    sheet.append([Rank,Name,Year,Rating])
```

```python
#data saved in excel
excel.save("Output_IMDb.xlsx")
```
